## 라이브러리 import

```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

## 웹 사이트 크롤링

```python
# 뉴스 카테고리
category = ['Politics', 'Economic', 'Social', 'Culture', 'World', 'IT']
```

```python
# 크롤링할 주소
url = 'https://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=100'
```

```python
# 차단 방지를 위해 유저 정보 추가
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.77 Safari/537.36'}
resp = requests.get(url, headers=headers)
```

```python
# 응답 결과 확인
print(type(resp))
print(list(resp))
```

```python
# 응답 결과 html 파싱
soup = BeautifulSoup(resp.text, 'html.parser')
print(soup)
```

```python
# bs 객체에서 선택자로 제목 태그 추출
title_tags = soup.select('.cluster_text_headline')
print(title_tags)
```

```python
# 태그에서 제목 텍스트만 추출
# 제목 텍스트에서 정규표현식에 해당하는 문자를 공백으로 변경
titles = [re.compile('[^가-힣|a-z|A-Z|0-9 ]+').sub(' ', title_tag.text) for title_tag in title_tags]
print(titles)
print(len(titles))
```

```python
# 크롤링한 페이지 저장할 데이터프레임
df_section_titles = pd.DataFrame()
url = f'https://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=100#&date=%2000:00:00&page='
for i in range(1,333):
    # url에 해당하는 웹 사이트 크롤링
    resp = requests.get(url + str(i), headers=headers)
    soup = BeautifulSoup(resp.text, 'html.parser')
    title_tags = soup.select('.cluster_text_headline')  # 기사 제목에 해당하는 class
    titles = [re.compile('[^가-힣|a-z|A-Z|0-9 ]+').sub(' ', title_tag.text) for title_tag in title_tags]  # 특수문자 제거
    
    # 크롤링 후 특수문자 제거한 데이터프레임
    df = pd.DataFrame(titles, columns=['title'])

    # 기존 인덱스를 무시하고 새로 생성해 결합
    df_section_titles = pd.concat([df_section_titles, df], ignore_index=True)
    
    # 진행상황 확인
    print(i, end=' ')
    if i % 50 == 0:
        print('\n')
```

## 함수로 만들어 6개 카테고리에 적용

```python
# 섹션 아이디와 이름, 페이지 수를 받아 크롤링한 데이터프레임을 반환하는 함수
def get_news_title(section_id, section_name, end_num):
    # 크롤링한 페이지 저장할 데이터프레임
    df_section_titles = pd.DataFrame()
    url = f'https://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1={section_id}#&date=%2000:00:00&page='

    for i in range(1, end_num + 1):
        # url에 해당하는 웹 사이트 크롤링
        resp = requests.get(url + str(i), headers=headers)
        soup = BeautifulSoup(resp.text, 'html.parser')
        title_tags = soup.select('.cluster_text_headline')  # 기사 제목에 해당하는 class
        titles = [re.compile('[^가-힣|a-z|A-Z|0-9 ]+').sub(' ', title_tag.text) for title_tag in title_tags]  # 특수문자 제거
        
        # 크롤링 후 특수문자 제거한 데이터프레임
        df = pd.DataFrame(titles, columns=['title'])
        
        # 기존 인덱스를 무시하고 새로 생성해 결합
        df_section_titles = pd.concat([df_section_titles, df], ignore_index=True)
        
        # 진행상황 확인
        print(i, len(df), '/', end=' ')
        if i % 50 == 0:
            print('\n')

    print('\n')
    df_section_titles['category'] = section_name
    return df_section_titles
```

```python
end_nums = [334, 423, 400, 87, 128, 74]  # 각 카테고리별 가져올 페이지 수
df_titles = pd.DataFrame()
for i, cat in enumerate(zip(category, end_nums)):
    i += 100
    print(i, cat[0], cat[1])  # 섹션 ID, 섹션 이름, 페이지 수
    df_titles = pd.concat([df_titles, get_news_title(i, cat[0], cat[1])])
```
